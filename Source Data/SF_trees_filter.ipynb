```python
# Dependencies
import json
import requests
import pandas as pd
import pprint as pp
from config import api_key
from config import app_token
import numpy as np
```

```python
# Open SF Tree CSV
source_file = "Source CSVs/SF_Trees.csv"

# Pull the source file into a dataframe
sourceTrees = pd.read_csv(source_file)
sourceTrees
```

# 2003  - 2007 Data

```python
# Just 2003 and more recent
filteredYear = sourceTrees[sourceTrees.Year >= 2003]
filteredYear
```

```python
# Just 2007 and earlier
filteredYear2007 = filteredYear[filteredYear.Year <= 2007]
filteredYear2007
```

```python
# Export to CSV
filteredYear2007.to_csv('Source CSVs/SF_Trees2003-2007.csv')
print("Export Complete")
```
