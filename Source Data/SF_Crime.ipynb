```python
# Endpoint for requesting SF tree data: https://data.sfgov.org/d/wg3w-h783
# Sample URL format for including an app token: https://data.seattle.gov/resource/3k2p-39jp.json?$$app_token=APP_TOKEN
```

```python
# Dependencies
import json
import requests
import pandas as pd
import pprint as pp
from config import api_key
from config import app_token
```

```python
#Set the base URL
baseURL = "https://data.sfgov.org/resource/tmnf-yvry.json"
```

```python
# Pull sample data

URL = baseURL + "?$limit=2210000&$offset=0" + "&$$app_token=" + app_token


# URL = baseURL + "?$$app_token=" + app_token
response = requests.get(URL).json()
print("Download complete")
# pp.pprint(response)
```

```python
len(response)
```

```python
#Convert JSON file to a data frame
sf_crime_df = pd.DataFrame(response)
sf_crime_df.head()
```

```python
# create a new data frame 
sf_crime = {} 
sf_crime = pd.DataFrame(sf_crime)
sf_crime[["Incident ID", "Category", "Description", "Day", "Date", "Time", "District", "Longitude", "Latitude"]] = sf_crime_df[["incidntnum", "category", "descript", "dayofweek", "date", "time", "pddistrict", "x", "y"]]
sf_crime.head(50)
```

```python
crime_sf = sf_crime.dropna()
crime_sf.head(50)
```

```python
cleanCrimeDate = crime_sf 
cleanCrimeDate['Date'] = crime_sf['Date'].map(lambda x: x.rstrip('T00:00:00.000'))
cleanCrimeDate
```

```python
# Pull Date Planted into a series
testSeries = cleanCrimeDate['Date']
```

```python
# Convert the hyphen in Date Planted into null
year = []
for i in testSeries:
    year.append(i.replace('-', ''))
year
```

```python
# Add the hyphen-less date back to the dataframe
cleanCrimeDate['IncidentYear'] = year
cleanCrimeDate.head()
```

```python
# Pull the year from PlantYear into a new column
cleanCrimeDate['Year'] = cleanCrimeDate['IncidentYear'].astype(str).str[0] + cleanCrimeDate['IncidentYear'].astype(str).str[1] + cleanCrimeDate['IncidentYear'].astype(str).str[2] + cleanCrimeDate['IncidentYear'].astype(str).str[3]
cleanCrimeDate
```

```python
cleanCrimeDate = cleanCrimeDate.drop(['Date', 'IncidentYear'], axis=1)
cleanCrimeDate
```

```python
# Output the data to CSV
output_path = "/Output CSVs/SF_Crime.csv"
cleanCrimeDate.to_csv("Source CSVs/SF_Crime.csv")
print("Output Complete")
```
